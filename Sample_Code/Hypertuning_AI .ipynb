```python
def load_real_time_data(symbol):
    """
    Fetch the most recent stock data for the given symbol.
    """
    print(f"Fetching real-time data for {symbol}...")
    # Fetch real-time data (current day and previous day)
    stock_data = yf.download(tickers=symbol, period="2d", interval="1m")
    stock_data.reset_index(inplace=True)
    stock_data.rename(columns={
        "Open": "open",
        "High": "high",
        "Low": "low",
        "Close": "close",
        "Volume": "volume"
    }, inplace=True)
    return stock_data
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

# Function to load stock data
def load_real_time_data(symbol):
    print(f"Fetching real-time data for {symbol}...")
    stock_data = yf.download(symbol, period="1d", interval="1m")  # 1-minute intervals for real-time data
    stock_data['intraday_avg'] = stock_data['Close'].rolling(window=5).mean()  # Example of intraday average
    return stock_data

# Real-time analysis function
def real_time_analysis(stock_data):
    """
    Analyze the most recent stock data for recommendations.
    """
    if stock_data.empty:
        print("No data available for analysis.")
        return "No Recommendation"
    
    print("Analyzing real-time data...")
    
    # Extract the most recent row
    latest_data = stock_data.iloc[-1]  # Get the most recent row
    
    # Access the scalar values from the last row using iloc
    close_price = latest_data["Close"]  # Scalar value
    intraday_avg = latest_data["intraday_avg"]  # Scalar value

    # Ensure close_price and intraday_avg are scalars
    close_price = close_price.item() if isinstance(close_price, pd.Series) else close_price
    intraday_avg = intraday_avg.item() if isinstance(intraday_avg, pd.Series) else intraday_avg

    # Comparison between scalar values
    if close_price > intraday_avg:
        recommendation = "Buy"
    else:
        recommendation = "Hold"
    
    print(f"Latest Close Price: {close_price}")
    print(f"Intraday Average: {intraday_avg}")
    print(f"Recommendation: {recommendation}")
    
    return recommendation

# Function to plot the Efficient Frontier
def plot_efficient_frontier(returns):
    num_assets = len(returns.columns)
    mean_returns = returns.mean()
    cov_matrix = returns.cov()
    
    # Simulate portfolio returns and volatilities
    port_returns = []
    port_volatility = []
    for _ in range(10000):
        weights = np.random.random(num_assets)
        weights /= np.sum(weights)
        
        port_returns.append(np.sum(weights * mean_returns))
        port_volatility.append(np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights))))
    
    # Convert lists to NumPy arrays
    port_returns = np.array(port_returns)
    port_volatility = np.array(port_volatility)
    
    # Plot the Efficient Frontier
    plt.figure(figsize=(10, 6))
    plt.scatter(port_volatility, port_returns, c=port_returns / port_volatility, cmap='viridis')
    plt.colorbar(label="Sharpe Ratio")
    plt.xlabel("Volatility (Risk)")
    plt.ylabel("Return")
    plt.title("Efficient Frontier")
    plt.show()

# Modern Portfolio Theory (MPT) - Calculate optimal portfolio weights
def modern_portfolio_theory(returns):
    num_assets = len(returns.columns)
    mean_returns = returns.mean()
    cov_matrix = returns.cov()
    
    # Simulate portfolio returns and volatilities
    port_returns = []
    port_volatility = []
    port_weights = []
    for _ in range(10000):
        weights = np.random.random(num_assets)
        weights /= np.sum(weights)
        
        port_returns.append(np.sum(weights * mean_returns))
        port_volatility.append(np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights))))
        port_weights.append(weights)
    
    # Convert lists to NumPy arrays
    port_returns = np.array(port_returns)
    port_volatility = np.array(port_volatility)
    
    # Find the portfolio with the highest Sharpe Ratio
    sharpe_ratios = port_returns / port_volatility
    max_sharpe_idx = np.argmax(sharpe_ratios)
    
    # Optimal weights
    optimal_weights = port_weights[max_sharpe_idx]
    
    return optimal_weights, port_returns, port_volatility

# Sharpe Ratio Optimization
def sharpe_ratio_optimization(returns):
    num_assets = len(returns.columns)
    mean_returns = returns.mean()
    cov_matrix = returns.cov()
    
    # Simulate portfolio returns and volatilities
    port_returns = []
    port_volatility = []
    sharpe_ratios = []
    for _ in range(10000):
        weights = np.random.random(num_assets)
        weights /= np.sum(weights)
        
        ret = np.sum(weights * mean_returns)
        vol = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
        
        port_returns.append(ret)
        port_volatility.append(vol)
        sharpe_ratios.append(ret / vol)
    
    # Convert lists to NumPy arrays
    port_returns = np.array(port_returns)
    port_volatility = np.array(port_volatility)
    sharpe_ratios = np.array(sharpe_ratios)
    
    # Find the portfolio with the maximum Sharpe ratio
    max_sharpe_idx = np.argmax(sharpe_ratios)
    
    return port_returns[max_sharpe_idx], port_volatility[max_sharpe_idx], sharpe_ratios[max_sharpe_idx]

# Neutral Strategy Model (Equal-weighted portfolio)
def neutral_strategy_model(returns):
    num_assets = len(returns.columns)
    neutral_weights = np.ones(num_assets) / num_assets  # Equal-weighted portfolio
    return neutral_weights

# Example usage
if __name__ == "__main__":
    # Load stock data for a symbol (e.g., "AAPL")
    symbol = "AAPL"
    stock_data = load_real_time_data(symbol)
    
    # Perform real-time analysis
    recommendation = real_time_analysis(stock_data)
    
    # Load historical returns for a portfolio (example: S&P 500, other stocks, etc.)
    symbols = ["AAPL", "GOOGL", "AMZN", "MSFT"]
    data = yf.download(symbols, start="2020-01-01", end="2024-01-01")["Adj Close"]
    returns = data.pct_change().dropna()
    
    # 1. Modern Portfolio Theory (MPT)
    optimal_weights, port_returns, port_volatility = modern_portfolio_theory(returns)
    print(f"Optimal Portfolio Weights (MPT): {optimal_weights}")
    
    # 2. Sharpe Ratio Optimization
    max_sharpe_return, max_sharpe_volatility, max_sharpe_ratio = sharpe_ratio_optimization(returns)
    print(f"Max Sharpe Ratio Portfolio Return: {max_sharpe_return}")
    print(f"Max Sharpe Ratio Portfolio Volatility: {max_sharpe_volatility}")
    print(f"Max Sharpe Ratio: {max_sharpe_ratio}")
    
    # 3. Neutral Strategy Model (Equal-weighted portfolio)
    neutral_weights = neutral_strategy_model(returns)
    print(f"Neutral Strategy Weights: {neutral_weights}")
    
    # Plot Efficient Frontier
    plot_efficient_frontier(returns)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer

# Function to load stock data
def load_real_time_data(symbol):
    print(f"Fetching real-time data for {symbol}...")
    stock_data = yf.download(symbol, period="1d", interval="1m")  # 1-minute intervals for real-time data
    stock_data['intraday_avg'] = stock_data['Close'].rolling(window=5).mean()  # Example of intraday average
    return stock_data

# Real-time analysis function
def real_time_analysis(stock_data):
    if stock_data.empty:
        print("No data available for analysis.")
        return "No Recommendation"
    
    print("Analyzing real-time data...")
    
    latest_data = stock_data.iloc[-1]
    close_price = latest_data["Close"]
    intraday_avg = latest_data["intraday_avg"]

    # Ensure close_price and intraday_avg are scalars
    close_price = close_price.item() if isinstance(close_price, pd.Series) else close_price
    intraday_avg = intraday_avg.item() if isinstance(intraday_avg, pd.Series) else intraday_avg

    if close_price > intraday_avg:
        recommendation = "Buy"
    else:
        recommendation = "Hold"
    
    print(f"Latest Close Price: {close_price}")
    print(f"Intraday Average: {intraday_avg}")
    print(f"Recommendation: {recommendation}")
    
    return recommendation

# Function to plot the Efficient Frontier
def plot_efficient_frontier(returns):
    num_assets = len(returns.columns)
    mean_returns = returns.mean()
    cov_matrix = returns.cov()
    
    # Simulate portfolio returns and volatilities
    port_returns = []
    port_volatility = []
    for _ in range(10000):
        weights = np.random.random(num_assets)
        weights /= np.sum(weights)
        
        port_returns.append(np.sum(weights * mean_returns))
        port_volatility.append(np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights))))
    
    port_returns = np.array(port_returns)
    port_volatility = np.array(port_volatility)
    
    # Plot the Efficient Frontier
    plt.figure(figsize=(10, 6))
    plt.scatter(port_volatility, port_returns, c=port_returns / port_volatility, cmap='viridis')
    plt.colorbar(label="Sharpe Ratio")
    plt.xlabel("Volatility (Risk)")
    plt.ylabel("Return")
    plt.title("Efficient Frontier")
    plt.show()

# Function to evaluate a portfolio
def portfolio_metrics(weights, mean_returns, cov_matrix):
    # Calculate portfolio return and volatility
    portfolio_return = np.sum(weights * mean_returns)
    portfolio_volatility = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
    
    return portfolio_return, portfolio_volatility

# Define scoring function for GridSearchCV
def sharpe_ratio_scoring(weights, mean_returns, cov_matrix):
    portfolio_return, portfolio_volatility = portfolio_metrics(weights, mean_returns, cov_matrix)
    sharpe_ratio = portfolio_return / portfolio_volatility
    return sharpe_ratio

# Modern Portfolio Theory (MPT) - Hyperparameter tuning for simulations
def modern_portfolio_theory_tuning(returns, n_simulations=10000):
    num_assets = len(returns.columns)
    mean_returns = returns.mean()
    cov_matrix = returns.cov()
    
    port_returns = []
    port_volatility = []
    port_weights = []
    
    for _ in range(n_simulations):
        weights = np.random.random(num_assets)
        weights /= np.sum(weights)
        
        port_returns.append(np.sum(weights * mean_returns))
        port_volatility.append(np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights))))
        port_weights.append(weights)
    
    port_returns = np.array(port_returns)
    port_volatility = np.array(port_volatility)
    
    # Find portfolio with the highest Sharpe Ratio
    sharpe_ratios = port_returns / port_volatility
    max_sharpe_idx = np.argmax(sharpe_ratios)
    
    return port_returns[max_sharpe_idx], port_volatility[max_sharpe_idx], port_weights[max_sharpe_idx]

# Sharpe Ratio Optimization - Hyperparameter tuning for simulations
def sharpe_ratio_optimization_tuning(returns, n_simulations=10000):
    num_assets = len(returns.columns)
    mean_returns = returns.mean()
    cov_matrix = returns.cov()
    
    port_returns = []
    port_volatility = []
    sharpe_ratios = []
    
    for _ in range(n_simulations):
        weights = np.random.random(num_assets)
        weights /= np.sum(weights)
        
        ret, vol = portfolio_metrics(weights, mean_returns, cov_matrix)
        port_returns.append(ret)
        port_volatility.append(vol)
        sharpe_ratios.append(ret / vol)
    
    port_returns = np.array(port_returns)
    port_volatility = np.array(port_volatility)
    sharpe_ratios = np.array(sharpe_ratios)
    
    # Find the portfolio with the maximum Sharpe ratio
    max_sharpe_idx = np.argmax(sharpe_ratios)
    
    return port_returns[max_sharpe_idx], port_volatility[max_sharpe_idx], sharpe_ratios[max_sharpe_idx]

# Neutral Strategy Model (Hyperparameter tuning for asset allocation)
def neutral_strategy_model_tuning(returns):
    num_assets = len(returns.columns)
    neutral_weights = np.ones(num_assets) / num_assets  # Equal-weighted portfolio
    return neutral_weights

# GridSearch or RandomSearch for Hyperparameter Tuning
def hyperparameter_tuning(returns):
    # Define parameter grid for tuning
    param_grid = {
        'n_simulations': [5000, 10000, 15000],
    }

    # Hyperparameter tuning for Modern Portfolio Theory
    best_sharpe_ratio = -np.inf
    best_params = None
    best_result = None
    
    for n_sim in param_grid['n_simulations']:
        mpt_result = modern_portfolio_theory_tuning(returns, n_simulations=n_sim)
        mpt_sharpe_ratio = mpt_result[0] / mpt_result[1]  # Sharpe ratio
        
        if mpt_sharpe_ratio > best_sharpe_ratio:
            best_sharpe_ratio = mpt_sharpe_ratio
            best_params = {'n_simulations': n_sim}
            best_result = mpt_result
    
    print(f"Best MPT result with {best_params['n_simulations']} simulations:")
    print(f"Optimal Portfolio Return: {best_result[0]}")
    print(f"Optimal Portfolio Volatility: {best_result[1]}")
    print(f"Optimal Weights: {best_result[2]}")
    
    return best_params, best_result

# Example usage
if __name__ == "__main__":
    # Load historical returns for a portfolio (example: S&P 500, other stocks, etc.)
    symbols = ["AAPL", "GOOGL", "AMZN", "MSFT"]
    data = yf.download(symbols, start="2020-01-01", end="2024-01-01")["Adj Close"]
    returns = data.pct_change().dropna()
    
    # Hyperparameter tuning
    best_params, best_result = hyperparameter_tuning(returns)
    
    # Plot Efficient Frontier (based on best simulation results)
    plot_efficient_frontier(returns)
```
